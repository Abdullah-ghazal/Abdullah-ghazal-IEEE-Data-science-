```python
###### part4:  Filtering - Using Conditionals to Filter Rows and Columns #########
# 1. Creating a Filter Mask
df_from_csv = pd.read_csv("Automobile_data.csv")
fltr = df_from_csv['make'] == 'audi'
print(fltr)
```

```python
#  2. Applying the Filter
print(df_from_csv[fltr])
```

```python
# 3. Selecting Specific Columns with .loc
print(df_from_csv.loc[fltr, ['make', 'aspiration']])
```

```python
# 4. Combining Conditions with & and |
fltr = (df_from_csv['fuel-type'] == 'gas') | (df_from_csv['drive-wheels'] == 'fwd')
print(df_from_csv.loc[fltr])
```

```python
#  5. Negating a Filter with ~

negated = ~(df_from_csv['make'] == 'audi')
print(df_from_csv.loc[negated])
```

```python
# 6. Filtering with String Methods: .str.contains()
fltr = df_from_csv['make'].str.contains('aud', na=False)
print(df_from_csv.loc[fltr])
```

```python
#### part5: Updating Rows and Columns - Modifying Data Within DataFrames #######

# 1. Updating Columns in a DataFrame
##  View Existing Column Names
df = df_from_csv
print(df.columns)
```

```python
##  Rename All Columns
df.column = ['Symboling', 'Normalized-Losses']
print(df.head())
```

```python
## Modify Column Names with String Methods

# Replace spaces with underscores
df.columns = df.columns.str.replace('-', '_')
print(df.columns)
```

```python
df.rename(columns={'symboling': 'Symboling', 'normalized_losses': 'Normalized_Nosses'}, inplace=True)

print(df.column)
```

```python
#  2. Updating Rows
## Update a Single Cell
df.loc[2,'make'] = 'BMW'

print(df)
```

```python
## Update Multiple Columns in One Row
df.loc[3,['fuel_type', 'num_of_doors']] = ['elec', 'four']

print(df)
```

```python
## Use Filter to Update Rows
fltr = (df['make'] == "audi")
df.loc[fltr, 'fuel_type'] = 'elec'

print(df)
```

```python
# 3.Advanced Methods: apply, map, applymap, replace
## .apply() on Series
df['make_length'] = df['make'].apply(len)

print(df)
```

```python
## .apply() with Custom Function
def make_upper(text):
    return text.upper()

df['make'] = df['make'].apply(make_upper)

print(df)
```

```python
## .applymap() on DataFrame

df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
print(df)
```

```python
## .map() on Series

df['num_of_doors'] = df['num_of_doors'].map({'four': 4, 'two': 2})
print(df)
```

```python
## .replace() on Series 

df['num_of_doors'] = df['num_of_doors'].replace({'four': 4, 'two': 2})
print(df)
```

```python
#### part6: Add/Remove Rows and Columns From DataFrames #######

# Adding Columns
## 1. Concatenating Columns to Form a New One


df['make_aspiration'] = df['make'] + ' ' + df['aspiration']
print(df)
```

```python
## 2. Column Splitting using .str.split()
df[['make' , 'aspiration']] = df['make_aspiration'].str.split(' ',  expand=True)

print(df)
```

```python
## Removing Columns
df.drop(columns=['aspiration'], inplace=True)
```

```python
## Adding Rows

new_row = {'Symboling ': '500', 'Normalized_Nosses': '32', 'make': 'Abiza'}
df = df.append(new_row, ignore_index=True)

```

```python
## Removing Rows
### a) by index
df.drop(index=0, inplace=True)  

### b) Conditionally
filt = (df['make'] == 'Abiza')
df.drop(index=df[filt].index, inplace=True)

print(df)
```
